```python
products = ["t-shirt", "mug", "hat", "book", "keychain"]

for product in products:
    print (product)
    print ("This product is available!")
```

```python
lista_compras_cliente = set ()
producto = input ("Producto que desea ordenar: ")

while True:
    if producto in products:
        lista_compras_cliente.add (producto)
        print ("Producto agregado a la lista de compras.")
    else:
        print ("El producto no está disponible.")
    
    continuar = input ("¿Desea agregar otro producto? (si/no): ")
    if continuar.lower() != 'si':
        break
    producto = input ("Producto que desea ordenar: ")

print (lista_compras_cliente)
```

```python
inventory = {
    "t-shirt": 10,
    "mug": 5,
    "hat": 0,
    "book": 2,
    "keychain": 15
}
for item in lista_compras_cliente:
    if inventory[item] > 0:
        print (f"El producto {item} está en stock.")
        inventory[item] -= 1
        print (f"Cantidad disponible: {inventory[item]}")
    else:
        print (f"El producto {item} no está en stock.")     
```
